```python
import seaborn as sns
%matplotlib inline
import pandas as pd
```

```python
prediction_rdm = pd.read_csv("hypothetical_model.csv", index_col=0)
```

```python
prediction_rdm.head()

```

```python
import matplotlib.pyplot as plt


# Figure with values filled in 

plt.figure(figsize = (10,10))
ax = sns.heatmap(prediction_rdm, cmap = 'viridis', annot=True, linecolor='black', linewidths=0.3, cbar=False, square=True,  annot_kws={'size': 15})                              
plt.tick_params(axis='both', which='major', labelsize=11, labelbottom = False, bottom=False, top = True, labeltop=True)
plt.yticks(rotation=0, weight='bold')
plt.xticks(rotation=90, weight='bold')
b, t = plt.ylim()
c,u = plt.xlim()
ax.vlines(x = [0,4,8,12], ymin = b, ymax = t, colors = 'black', lw = 3)
ax.hlines(y=[0,4,8,12], xmin = c, xmax = u, colors = 'black', lw = 3)
plt.savefig("hypothetical_RDM.png", transparent = True, dpi=400)
```

```python
# Figure with no values (for method 2):


plt.figure(figsize = (10,10), edgecolor='black')
ax = sns.heatmap(prediction_rdm, cmap = 'viridis', annot=False, cbar=False,linecolor='black', linewidths=0.3, square=True,  annot_kws={'size': 15})                              
plt.tick_params(axis='both', which='major', labelsize=11, labelbottom = False, bottom=False, top = True, labeltop=True)
plt.yticks(rotation=0, weight='bold')
plt.xticks(rotation=90, weight='bold')
b, t = plt.ylim()
c,u = plt.xlim()
ax.vlines(x = [0,12], ymin = b, ymax = t, colors = 'black', lw = 2)
ax.hlines(y=[0,12], xmin = c, xmax = u, colors = 'black', lw = 2)
plt.savefig("model_ondm.png", transparent = True, dpi=400)

```
