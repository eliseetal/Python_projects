```python
# Importing libraries
import numpy as np
import pandas as pd
from scipy.stats import mode
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
 
%matplotlib inline
```

```python
# Drop missing values

data = pd.read_csv("export.csv").dropna()

missing_values = data.isnull().mean()
missing_values  # There are no missing values now that we have removed them. 
```

```python
data = data.assign(age = lambda x: x.age.where(x.age.ge(0))) # ge = greater than or equal to
data = data.assign(age = lambda x: x.age.where(x.age.le(116)))  # le = less than or equal to 

print("The maximum age in our dataset is:", data['age'].max())
print("The minimum age in our dataset is:", data['age'].min())
```

```python
# Let's create a dataframe for the symptoms 

symptoms = data[data['trackable_type'] == 'Symptom']
top500symptoms = data[data['trackable_type'] == 'Symptom'].trackable_name.value_counts().iloc[0:500]
# symptoms['trackable_value'] = pd.to_numeric(symptoms['trackable_value']) # in case we want to analyse the trackable value (bu twe don't know what that is from the description) 

# and a dataframe for the conditions 

conditions = data[data['trackable_type'] == 'Condition']
top20conditions = data[data['trackable_type'] == 'Condition'].trackable_name.value_counts().iloc[0:20]
# conditions['trackable_value'] = pd.to_numeric(conditions['trackable_value'])
```

```python
# Top 10 symptoms by number of reports and top 10 conditions reported on the app. 

print(top500symptoms.head(10))
print('\n')
print(top20conditions.head(10))
```

```python
top_500= symptoms[symptoms['trackable_name'].isin(top500symptoms.index)]



top_500.head()

row_indexer = top_500.index 
col_indexer = 'symptom_present'
value = 1

top_500.loc[row_indexer, col_indexer] = value

```

```python
# change from long to wide format 

top_500_symptoms_pivot = top_500.pivot_table(columns = 'trackable_name', index=['user_id', 'checkin_date'], values='symptom_present').reset_index()

```

```python
top_500_symptoms_pivot.head()
```

```python
top_500_symptoms_pivot = top_500_symptoms_pivot.fillna(0)


top_500_symptoms_pivot.head()
```

```python
# only keep the top 20 conditions

top_20_conditions= conditions[conditions['trackable_name'].isin(top20conditions.index)]

top_20_conditions.head()
```

```python
df = pd.merge(top_500_symptoms_pivot, top_20_conditions, on=["user_id", "checkin_date"])
```

```python
df.head(1000)
```

```python
# Checking whether the dataset is balanced or not
condition_counts = df["trackable_name"].value_counts()
temp_df = pd.DataFrame({
    "Condition": condition_counts.index,
    "Counts": condition_counts.values
})
 
plt.figure(figsize = (18,8))
sns.barplot(x = "Condition", y = "Counts", data = temp_df)
plt.xticks(rotation=90)
plt.show()

```

```python
# features
X = df.drop(['trackable_name', 'user_id', 'checkin_date', 'age', 'sex', 'country', 'trackable_id', 'trackable_type', 'trackable_value'],axis=1)


from sklearn.preprocessing import LabelEncoder
# give each condition a unique value 
encoder = LabelEncoder()
df['trackable_name'] = encoder.fit_transform(df['trackable_name'])


y = df['trackable_name']
```

```python
from imblearn.under_sampling import RandomUnderSampler
# from imblearn.over_sampling import RandomOverSampler

# Create an instance of RandomOverSampler
sampler = RandomUnderSampler(random_state=42)

# Apply undersampling to your feature matrix X and target variable y
X_resampled, y_resampled = sampler.fit_resample(X, y)
```

```python
rcondition_counts = y_resampled.value_counts()
temp_df = pd.DataFrame({
    "Condition": rcondition_counts.index,
    "Counts": rcondition_counts.values
})
 
plt.figure(figsize = (18,8))
sns.barplot(x = "Condition", y = "Counts", data = temp_df)
plt.xticks(rotation=90)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_resampled,y_resampled,test_size=0.30, random_state = 101)


print(f"Train: {X_train.shape}, {y_train.shape}")
print(f"Test: {X_test.shape}, {y_test.shape}")
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=100)
rfc.fit(X_train, y_train)
```

```python
rfc_pred = rfc.predict(X_test)
```

```python
importances = rfc.feature_importances_

sorted_indices = importances.argsort()[::-1]
sorted_importances = importances[sorted_indices]
sorted_features = X.iloc[:,sorted_indices]

sorted_features.iloc[:,0:50].columns
```

```python
# colours of bars 

# Create a colourmap with 30 colors
cmap = plt.get_cmap('rainbow')


# top30 = sorted_features.iloc[:]

# Generate the colour codes for each bar
colours = [cmap(i) for i in np.linspace(0, 1, 50)]

colours = colours[::-1]


plt.figure(figsize=(20,20))
plt.bar(range(50), sorted_importances[0:50], color=colours)
plt.xticks(range(50), sorted_features.iloc[:, 0:50], rotation=90)
plt.xlabel('Features')
plt.ylabel('Importance')
plt.title('Feature Importance')
# plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.model_selection import cross_val_score
```

```python
print(confusion_matrix(y_test,rfc_pred))
print('\n')
print(classification_report(y_test,rfc_pred))
report_rfc = classification_report(y_test,rfc_pred, output_dict=True)

rfc_macro_f1 = report_rfc['macro avg']['f1-score']

print("%s: %f " % ('RFC', rfc_macro_f1))
```
