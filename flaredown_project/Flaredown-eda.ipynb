### Clean the data

```python
missing_values = data.isnull().mean()
missing_values
```

```python
#### We can see that we have some missing values. For now, we will not remove these. 
```

```python
data.describe()
```

```python
# We can see that the ranges for the age seems a bit odd. The minimum age is negative, and the maximum age is 2018. Only vampires are that old. 
# So let's remove those ages that seem incorrect. 
# The oldest living person is currently 116, so we will use this for our upper limit, and we will set 0 as our lower limit. 

# data['age'] = data['age'][(data['age'] < 117)]
# data['age'] = data['age'][(data['age'] > 0)]  # this will remove people that are reported to be 0 years old though.

# using lambda expressions

data = data.assign(age = lambda x: x.age.where(x.age.ge(0))) # ge = greater than or equal to
data = data.assign(age = lambda x: x.age.where(x.age.le(116)))  #Â le = less than or equal to 

# translation of the above: assign a new column to our data (data.assign), called "age", which takes in the data.age column, and assigns it the age, but only when the age is greater than or equal to 0. 

print("The maximum age in our dataset is:", data['age'].max())
print("The minimum age in our dataset is:", data['age'].min())



# Let's also drop NaN values because we want to only include rows that have a value in the "trackable value" column 

data = data.dropna(axis=0)
```

### Let's look at the top 25 conditions and get the average based on the mean "trackable value" per condition

#### Let's run a linear mixed effects model to try to predict the effect of symptom severity on overall reported Fibromyalgia severity 




```python
import researchpy as rp
import statsmodels.api as sm
import statsmodels.formula.api as smf
import scipy.stats as stats
from statsmodels.tools.sm_exceptions import ConvergenceWarning
```

```python
# Using the top 10 reported symptoms on Flaredown for Fibromyalgia 

md = smf.mixedlm("Fibromyalgia~C(Fatigue)+C(Headache)+C(Nausea)+C(Brain_fog)+C(Joint_pain)+C(Fatigue_and_tiredness)+C(Anxiety)+C(Neck_pain)+C(Dizziness)+C(Back_pain)", data=joined, groups=joined["user_id"])
# md = smf.mixedlm("Weight ~ Time", data, groups=data["Pig"])

mdf = md.fit()

print(mdf.summary())
#  md = smf.mixedlm("Weight ~ Time", data, groups=data["Pig"])


predictions = mdf.predict()
residuals = mdf.resid
```

### Insights

The coefficients for the categorical fixed effects represent the expected change in the log-odds of Fibromyalgia severity when comparing each level to the reference level. Let's look at a few examples:

Fatigue: The coefficients for the different levels of Fatigue indicate how the log-odds of Fibromyalgia severity change as the severity of fatigue increases. For example, an increase from level 0 to level 1 of Fatigue is associated with a decrease of 0.271 in the log-odds, while an increase from level 0 to level 4 is associated with an increase of 0.503 in the log-odds.

Headache: The coefficients for the different levels of Headache indicate how the log-odds of Fibromyalgia severity (as rated by the user) change as the severity of headache increases. For example, an increase from level 0 to level 3 of Headache is associated with an increase of 0.127 in the log-odds.

Nausea: The coefficients for the different levels of Nausea indicate how the log-odds of having Fibromyalgia change as the severity of nausea increases. For example, an increase from level 0 to level 4 of Nausea is associated with an increase of 0.131 in the log-odds.


Group Var: The estimated variance component for the random effects (Group Var) represents the variability in the random intercepts across different groups. It indicates the extent to which the log-odds of Fibromyalgia severity vary between different groups in your data. In this case, the estimated value is 2.215.
